```python
import funcs
import mrsobs

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
plt.style.use('presentation')
%matplotlib notebook

import warnings
warnings.simplefilter('ignore')
```

```python
# Define paths to data
workDir   = '/Users/ioannisa/Desktop/python/miri_devel/'
cdpDir    = workDir+'cdp_data/CDP7/'
```

```python
# load objects
sliceMap_file = fits.open(cdpDir+'MIRI_FM_MIRIFUSHORT_12SHORT_SLICEMAP_7B.00.00.fits')

slice_map = {}
extension = 0
for transmission in ['10','20','30','40','50','60','70','80','90']:
    extension += 1
    slice_map[transmission] = sliceMap_file[extension].data
    
    plt.figure(figsize=(6,6))
    plt.title('Slice map band 1A/2A, {}% transmission'.format(transmission),fontsize=16)
    plt.imshow(slice_map[transmission])
    plt.xlabel('X-direction [pix]',fontsize=16)
    plt.ylabel('Y-direction [pix]',fontsize=16)
    plt.tick_params(axis='both',labelsize=16)
    plt.tight_layout()
```
